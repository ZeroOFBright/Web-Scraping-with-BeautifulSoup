```python
import pandas as pd
import time
```

# Scrap MP from PubChem

```python
def web_scrap_mp(dataframe):
  df = dataframe.copy()
  driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
  for i in range(len(df)):
    cid = str(df.loc[i,"cid"])
    url = "https://pubchem.ncbi.nlm.nih.gov/compound/"+cid+"#section=Melting-Point"
    driver.get(url)
    time.sleep(1)
    
    content = driver.page_source

    soup = BeautifulSoup(content)

    mp = []

    find_melting_point = soup.find(id="Melting-Point")

    if(find_melting_point != None):
      melting_point = find_melting_point.findAll('div', attrs={'class':'section-content-item'})

      for value in melting_point:
        x = value.find("p")
        mp.append(x.text)

      df.loc[i,"mp"] = str(mp) 
  return df
```

```python
import ast
import numpy as np
def clean_f(dataframe):
  df = dataframe.copy()
  for i in range (len(df)):
    try :
      li = list(set(ast.literal_eval(df.loc[i,"mp"])))
      tem_c = []
      for temperature in li :
        if "°C" in temperature :
          tem_c.append(temperature)
      df.loc[i,"mp"] = str(tem_c)
    except :
      continue
  return df
```

# Scrap smile

```python
def web_scrap_smiles(dataframe):
  df = dataframe.copy()
  driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
  for i in range(len(df)):
    cid = str(df.loc[i,"cid"])
    url = "https://pubchem.ncbi.nlm.nih.gov/compound/"+cid
    driver.get(url)
    time.sleep(1)
    
    content = driver.page_source

    soup = BeautifulSoup(content)

  
    find_smile = soup.find(id="Canonical-SMILES")

    if(find_smile != None):
      smile = find_smile.find('div', attrs={'class':'section-content-item'}).find("p")
      df.loc[i,"smiles"] = smile.text 

  return df
```
